# Area theorem (GW150914)

```python
import numpy
import matplotlib.pyplot as plt
```

```python
delA_NRSur_hmeco_NRSur_3ms = numpy.loadtxt("data/GW1501914/DeltaAbyA_gatestart_hmeco_gateend_3ms_NRSur")
delA_NRSur_0ms_NRSur_3ms = numpy.loadtxt("data/GW1501914/DeltaAbyA_gatestart_tm_gateend_3ms_NRSur")
delA_NRSur_0ms_NRSur_0ms = numpy.loadtxt("data/GW1501914/DeltaAbyA_gatestart_tm_gateend_0ms_NRSur")
delA_NRSur_hmeco_Ring_3ms_220 = numpy.loadtxt("data/GW1501914/DeltaAbyA_gatestart_hmeco_NRSur_gateend_3ms_220")
delA_NRSur_0ms_Ring_3ms_220 = numpy.loadtxt("data/GW1501914/DeltaAbyA_gatestart_tm_NRSur_gateend_3ms_220")
delA_NRSur_0ms_Ring_0ms = numpy.loadtxt("data/GW1501914/DeltaAbyA_gatestart_tm_NRSur_gateend_0ms_220-221")
delA_NRSur_0ms_Ring_0ms_220 = numpy.loadtxt("data/GW1501914/DeltaAbyA_gatestart_tm_NRSur_gateend_0ms_220")
```

```python
#### -----Figure 3 of arXiv:2111.13664 Left panel------

fig, (ax1) = plt.subplots(nrows = 1, ncols = 1, figsize=(6, 4))


hist1, bins1 = numpy.histogram(delA_NRSur_hmeco_NRSur_3ms, bins = 500, density=True)
center1 = bins1[:-1]
ax1.plot(center1, hist1, alpha = 1, color='#0096FF', ls="-.", label='NRSur7dq(before $t_{hmeco}$) - NRSur7dq(after $t_m+$3ms)')


hist3, bins3 = numpy.histogram(delA_NRSur_0ms_NRSur_3ms, bins = 500, density=True)
center3 = bins3[:-1]
ax1.plot(center3, hist3, alpha = 1, color='#8B3E2F', ls="-.", label='NRSur7dq(before $t_m$)-NRSur7dq(after $t_m+$3ms)')


hist2, bins2 = numpy.histogram(delA_NRSur_0ms_NRSur_0ms, bins = 500, density=True)
center2 = bins2[:-1]
ax1.plot(center2, hist2, alpha = 1, color='#FFC300', ls="--", label='NRSur7dq(before $t_m$)-NRSur7dq(after $t_m$)')



hist4, bins4 = numpy.histogram(delA_NRSur_hmeco_Ring_3ms_220, bins = 500, density=True)
center4 = bins4[:-1]
ax1.plot(center4, hist4, alpha = 1, color='k', ls="--", label='NRSur7dq(before $t_{hmeco}$) - Damped Sinusoid[220](after $t_m+$3ms)')

hist5, bins5 = numpy.histogram(delA_NRSur_0ms_Ring_3ms_220, bins = 500, density=True)
center5 = bins5[:-1]
ax1.plot(center5, hist5, alpha = 1, color='navy', ls="-.", label='NRSur7dq(before $t_m$)-Damped Sinusoid[220](after $t_m+$3ms)')


hist6, bins6 = numpy.histogram(delA_NRSur_0ms_Ring_0ms, bins = 500, density=True)
center6 = bins6[:-1]
ax1.plot(center6, hist6, alpha = 1, color='green', ls="--", label='NRSur7dq(before $t_m$)-Damped Sinusoid[220+221](after $t_m$)')


hist7, bins7 = numpy.histogram(delA_NRSur_0ms_Ring_0ms_220, bins = 500, density=True)
center7 = bins7[:-1]
ax1.plot(center7, hist7, alpha = 1, color='r', ls="-.", label='NRSur7dq(before $t_m$)-Damped Sinusoid[220](after $t_m$)')




ax1.axvspan(-2, 0, alpha=0.2, color='k')
ax1.legend(loc='upper center', prop={'size': 9.8})
ax1.set_xlabel('$\Delta A/A_i$')
ax1.set_ylabel('Probability density')
ax1.set_xlim(-1.2,2.5)
ax1.set_ylim(0,2.9)
fig.set_dpi(350)
plt.tight_layout()
```
